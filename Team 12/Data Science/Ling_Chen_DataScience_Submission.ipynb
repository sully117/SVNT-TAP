## Part 1: Data Exploration

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import seaborn as sns
import numpy as np
import itertools
# import cPickle
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
df_us = pd.read_csv('youtube-new/USvideos.csv',encoding ='utf-8',engine='python',sep=',', quotechar='"', error_bad_lines=False)
```

```python
def df_from_csv(path):
    df = pd.read_csv(path, nrows=1) # read just first line for columns
    columns = df.columns.tolist() # get the columns
    cols_to_use = columns[:len(columns)-1] # drop the last one
    df = pd.read_csv(path, usecols=cols_to_use)
    return df
```

```python
df = df_from_csv('/Users/chenling/Desktop/youtube-new/USvideos.csv')
```

```python
df.describe()
```

```python
# unique value overview
for column in df.columns:
    uniques = df[column].unique()
    print('{0:20s} {1:5d}\t'.format(column, len(uniques)), uniques[:2])
```

```python
df.likes.fillna(0,inplace=True)
# df.likes.notnull().astype('int')
df.dislikes.fillna(0,inplace=True)
# df.dislikes.notnull().astype('int')
df.views.fillna(0,inplace=True)
# df.views.notnull().astype('int')
df.comment_count.fillna(0,inplace=True)
# df.comment_count.notnull().astype('int')
```

```python
df['category_name'] = np.nan

df.loc[(df["category_id"] == 1),"category_name"] = 'Film and Animation'
df.loc[(df["category_id"] == 2),"category_name"] = 'Cars and Vehicles'
df.loc[(df["category_id"] == 10),"category_name"] = 'Music'
df.loc[(df["category_id"] == 15),"category_name"] = 'Pets and Animals'
df.loc[(df["category_id"] == 17),"category_name"] = 'Sport'
df.loc[(df["category_id"] == 19),"category_name"] = 'Travel and Events'
df.loc[(df["category_id"] == 20),"category_name"] = 'Gaming'
df.loc[(df["category_id"] == 22),"category_name"] = 'People and Blogs'
df.loc[(df["category_id"] == 23),"category_name"] = 'Comedy'
df.loc[(df["category_id"] == 24),"category_name"] = 'Entertainment'
df.loc[(df["category_id"] == 25),"category_name"] = 'News and Politics'
df.loc[(df["category_id"] == 26),"category_name"] = 'How to and Style'
df.loc[(df["category_id"] == 27),"category_name"] = 'Education'
df.loc[(df["category_id"] == 28),"category_name"] = 'Science and Technology'
df.loc[(df["category_id"] == 29),"category_name"] = 'Non Profits and Activism'
df.loc[(df["category_id"] == 25),"category_name"] = 'News & Politics'
```

```python
df['like_ratio'] =  df['likes'] / df['views'].astype(int) * 100
df['dislike_ratio'] =  df['dislikes'] / df['views'] * 100
df['comment_ratio'] =  df['comment_count'] / df['views'].astype(int) * 100
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20, 6))
sns.barplot(x='category_name', y='dislike_ratio', data=df, ax=ax);
ax.set_title('Mean dislike ratio per category', fontsize=16)
# plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20, 6))
sns.barplot(x='category_name', y='like_ratio', data=df, ax=ax);
ax.set_title('Mean like ratio per category', fontsize=16)
# plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20, 6))
sns.barplot(x='category_name', y='comment_ratio', data=df, ax=ax);
ax.set_title('Mean comment ratio per category', fontsize=16)
# plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (9,6))

g1 = sns.distplot(df['dislike_ratio'], color='red',hist=False, label="Dislike")
g1 = sns.distplot(df['like_ratio'], color='green',hist=False, label="Like")
g1 = sns.distplot(df['comment_ratio'],color='blue',hist=False,label="Comment")
g1.set_title('CONVERT RATE DISTRIBUITION', fontsize=16)
plt.legend()
plt.show()
```

For different catrgory group, the like/ dislike/ comment rate is diffrent from the barchart.

We can see that the Dislike rate is very low in almost all categories but some outliers in 'Pprofits and activism' and News & Politics.

The mean of ike distribuition is less than 5% but in music we have a very interesting pattern of like rate. The music category have the highest engagement rate, however, people seldom make comment under music videos and that makes sense because people just enjoy the videos.

And at the Comment rate we can see the gaming, how-to-and style, news and politics as weel as profit category have the highest rates of commments.
